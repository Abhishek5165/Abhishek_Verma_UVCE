# Merge the Datasets ü§ù

```python
# Merge Transactions with Customers and Products

transactions_customers = pd.merge(transactions, customers, on="CustomerID", how="inner")
merged_data = pd.merge(transactions_customers, products, on="ProductID", how="inner")

# Display the merged dataset
print("Merged Dataset Info:\n")

merged_data.info()
print("\n")
```

```python
# Convert date columns to datetime format

customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Check for duplicates

print("Duplicate rows in Customers:", customers.duplicated().sum())
print("Duplicate rows in Products:", products.duplicated().sum())
print("Duplicate rows in Transactions:", transactions.duplicated().sum())
```

# Exploratory Data Analysis (EDA) üëá

## Relationship analysis between Variables

```python
## Analyze Customer Distribution by Region

region_counts = customers['Region'].value_counts()
print("\nCustomer Distribution by Region:\n",region_counts)
```

## Regional Analysis 

```python
## Visualization: Customer Distribution by Region (bar plot) - 

colors = sns.color_palette("magma", len(region_counts))
plt.figure(figsize=(8, 5))
plt.bar(region_counts.index, region_counts.values, color=colors)
plt.title("Customer Distribution by Region", fontsize=14)
plt.xlabel("\nRegion", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.show()
```

```python
# Convert date columns to datetime format

merged_data['SignupDate'] = pd.to_datetime(merged_data['SignupDate'])
```

```python
## Analyze Signup Trends Over Time

merged_data['YearMonth'] = merged_data['SignupDate'].dt.to_period('M')
signup_trends = merged_data['YearMonth'].value_counts().sort_index()
print("\nSignup Trends Over Time (First 5 Rows):\n", signup_trends.head())
```

```python
## Visualization: Signup Trends Over Time

plt.figure(figsize=(10, 6))
signup_trends.plot(kind='line', marker='o', color='r', linewidth=2)
plt.title("Customer Signup Trends Over Time", fontsize=14)
plt.xlabel("Year-Month", fontsize=12)
plt.ylabel("Number of Signups", fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

```python
## Analyze Monthly Signup Trends

monthly_signups = merged_data['SignupDate'].dt.month.value_counts().sort_index()
print("\nMonthly Signup Trends:\n", monthly_signups)
```

## Monthly Trends

```python
# Visualization: Monthly Signup Trends

colors = sns.color_palette("viridis", len(monthly_signups))

plt.figure(figsize=(8, 5))
plt.bar(range(len(monthly_signups)), monthly_signups.values, color=colors)
plt.xticks(range(len(monthly_signups)), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)

# Add titles and labels
plt.title("Monthly Signup Trends", fontsize=14)
plt.xlabel("\nMonth", fontsize=12)
plt.ylabel("Number of Signups", fontsize=12)

plt.show()
```

```python
## Analyze Yearly Signup Growth

merged_data['Year'] = merged_data['SignupDate'].dt.year
yearly_growth = merged_data['Year'].value_counts().sort_index()
print("\nYearly Signup Growth:\n", yearly_growth)
```

## Yearly Trends

```python
## Visualization: Yearly Signup Trends

custom_colors = ["black","coral","purple"]

plt.figure(figsize=(8, 5))
plt.bar(range(len(yearly_growth)), yearly_growth.values, color=custom_colors) 
plt.xticks(range(len(yearly_growth)), yearly_growth.index) 

## Add titles and labels
plt.title("Yearly Signup Trends", fontsize=14)
plt.xlabel("\nYear", fontsize=12)
plt.ylabel("Number of Signups", fontsize=12)
plt.show()
```

```python

# Distribution of Price

plt.figure(figsize=(8, 5))
sns.histplot(products['Price'], bins=30, kde=True, color='blue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Product Category Analysis 

```python
# Transaction count and sales by category

category_counts = merged_data['Category'].value_counts()
category_sales = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)

# Plot transaction count by category

custom_palette = sns.color_palette(["grey", "red","blue","green"])

plt.figure(figsize=(8, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, hue=category_counts.index, palette=custom_palette)
plt.title("\nTransaction Count by Product Category", fontsize=16)
plt.xlabel("\nCategory", fontsize=12)
plt.ylabel("Transaction Count", fontsize=12)
plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()

# Plot total sales by category

plt.figure(figsize=(8, 6))
sns.barplot(x=category_sales.index, y=category_sales.values, hue=category_sales.index, palette=custom_palette)
plt.title("\nTotal Sales by Product Category", fontsize=16)
plt.xlabel("\nCategory", fontsize=12)
plt.ylabel("Total Sales Value", fontsize=12)
plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()
```

## Monthly Transactions Trends

```python
# Customer distribution by region

region_counts = merged_data['Region'].value_counts()

# Generate a color palette
colors = sns.color_palette("cividis", len(region_counts))

# Plot regional distribution
plt.figure(figsize=(8, 6))
plt.bar(region_counts.index, region_counts.values, color=colors) 
plt.title("Customer Distribution by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Convert 'TransactionDate' to datetime format
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'], errors='coerce')

# Check for any invalid date conversions
invalid_dates = merged_data['TransactionDate'].isna().sum()
print(f"Number of invalid dates: {invalid_dates}")

# Extract month from TransactionDate
merged_data['TransactionMonth'] = merged_data['TransactionDate'].dt.to_period('M')

# Monthly transaction count and total sales
monthly_counts = merged_data['TransactionMonth'].value_counts().sort_index()
monthly_sales = merged_data.groupby('TransactionMonth')['TotalValue'].sum()

# Plot monthly transaction count
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='blue', alpha=0.7)
plt.title("Monthly Transaction Count", fontsize=16)
plt.xlabel("\nMonth", fontsize=12)
plt.ylabel("Transaction Count", fontsize=12)
plt.tight_layout()
plt.show()

# Plot monthly total sales
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='orange', alpha=0.7)
plt.title("Monthly Total Sales", fontsize=16)
plt.xlabel("\nMonth", fontsize=12)
plt.ylabel("Total Sales Value", fontsize=12)
plt.tight_layout()
plt.show()
```
